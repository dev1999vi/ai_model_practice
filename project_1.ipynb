```python
#Word-level Tokenization

corpus = ["The quick brown fox jumps over the lazy dog.",
  "A journey of a thousand miles begins with a single step.",
  "All that glitters is not gold.",
  "The early bird catches the worm.",
  "You can't teach an old dog new tricks.",
  "Actions speak louder than words.",
  "Don't count your chickens before they hatch.",
  "The pen is mightier than the sword.",
  "When in Rome, do as the Romans do.",
  "The grass is always greener on the other side.",
  "Time flies when you're having fun.",
  "Better late than never, but never late is better.",
  "Two wrongs don't make a right.",
  "The apple doesn't fall far from the tree.",
  "Curiosity killed the cat, but satisfaction brought it back.",
  "You can lead a horse to water, but you can't make it drink.",
  "Don't put all your eggs in one basket.",
  "A picture is worth a thousand words.",
  "There's no place like home.",
  "The best things in life are free."]

PAD, UNK = "[PAD]", "[UNK]"

vocab = []
word2id = {}
id2word = {}

word_set = set()

for sen in corpus:
    words = sen.split()
    for word in words:
        word_set.add(word.lower())

vocab = [PAD, UNK] + list(word_set)

print(f"First 15 words in vocab: {vocab[:15]}")

for i, w in enumerate(vocab):
    word2id[w] = i
    id2word[i] = w

print(f'Vocabulary Size: {len(vocab)} words')

def encode(text):
    # output = []
    # for w in text.lower().split():
    #     if w in word2id:    
    #         output.append(word2id[w])
    #     else:
    #         output.append(word2id[UNK])
    # return output
    return [word2id.get(w, word2id[UNK]) for w in text.lower().split()]

def decode(arr):
    # output = ""
    # for i in arr:
    #     output+= f" {id2word[i]}"
    # return output
    return " ".join(id2word.get(i) for i in arr if i!=word2id[PAD])

    
print(encode("the make boy"))
print(decode([110,2, 1, 0]))
```

```python
#Character level Tokenizer
import string
char_lvl_vocab = ["[UNK]", "[PAD]"," "] + list(string.ascii_lowercase + string.ascii_uppercase)
char_lvl_word2id = {}
char_lvl_id2word = {}

char_lvl_id2word = { idx:ch for idx, ch in enumerate(char_lvl_vocab)}
char_lvl_word2id = { ch:idx for idx, ch in enumerate(char_lvl_vocab)}

# for i,c in enumerate(char_lvl_vocab):
#     char_lvl_word2id[c]=i
#     char_lvl_id2word[i]=c

def char_encode(text):
    return [char_lvl_word2id.get(w, char_lvl_word2id[UNK]) for w in text]

def char_decode(arr):
    return "".join(char_lvl_id2word.get(i) for i in arr if i!=char_lvl_word2id[PAD])

print(char_encode("Hello World!"))
print(char_decode([36, 7, 14, 14, 17, 2, 51, 17, 20, 14, 6, 0]))

```
